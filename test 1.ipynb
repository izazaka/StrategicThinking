```python
import pandas as pd

# Assuming you have a DataFrame df with a column named 'Description'
df = pd.DataFrame(["comment"])

# Function to detect language
def detect_language(text):
    # You can use a language detection library, such as langdetect
    from langdetect import detect
    try:
        lang = detect(text)
        return lang
    except:
        return 'unknown'

# Apply language detection to each row
df['Language'] = df['comment'].apply(detect_language)

# Separate into different columns based on language
df_english = df[df['Language'] == 'en']
df_portuguese = df[df['Language'] == 'pt']

# Drop the 'Language' column if you no longer need it
df_english = df_english.drop(columns=['Language'])
df_portuguese = df_portuguese.drop(columns=['Language'])

# Display the results
print("English Texts:")
print(df_english['Description'].values)

print("\nPortuguese Texts:")
print(df_portuguese['Description'].values)
```
