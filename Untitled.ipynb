```python
import pandas as pd
import numpy as np

# Assuming you have a DataFrame named 'f' with a column named 'English_Text'
# Replace the sample data with your actual DataFrame

data = {'English_Text': f["English_Text"].tolist()}
df = pd.DataFrame(data)

# Function to clean the text
def clean_text(text):
    if pd.notna(text):  # Check for NaN values
        # Remove "(Translated by Google)"
        text = text.replace('(Translated by Google)', '')

        # Remove "\n\n(Original)\n"
        text = text.replace('\n\n(Original)\n', '')

        # Remove "\" 
        text = text.replace('\\', '')

        return text.strip()  # Remove leading and trailing whitespaces
    else:
        return text  # Return NaN if the input is NaN

# Apply the cleaning function to each row in the 'English_Text' column
df['cleaned_text'] = df['English_Text'].apply(clean_text)

# Display the DataFrame with the cleaned text
print(df)
```
