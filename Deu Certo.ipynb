```python
import json
import pandas as pd
import glob

# Specify the path to the directory containing your JSON files
json_files_path = '/Users/izaiaszacca/Documents/Reviews/*.json'
#/Users/izaiaszacca/Documents/Reviews
#/Users/izaiaszacca/Documents/Reviews/GoogleReviews json

# Create an empty list to store DataFrames
dfs = []

# Iterate over each JSON file in the specified directory
for json_file_path in glob.glob(json_files_path):
    with open(json_file_path, 'r') as file:
        data = json.load(file)

    # Convert the JSON data to a Pandas DataFrame
    df = pd.DataFrame(data)

    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Specify the path for the combined CSV file
csv_output_path = '/Users/izaiaszacca/Documents/Reviews/gr.csv'

# Save the combined DataFrame to a CSV file
combined_df.to_csv(csv_output_path, index=False)

print(f'Combined data has been saved to {csv_output_path}')
```

```python
import pandas as pd
import json

# Specify the path to your CSV file
csv_file_path = '/Users/izaiaszacca/Documents/Reviews/gr.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Define a function to extract information from the JSON strings
def extract_info(json_string, key):
    try:
        # Convert the JSON string to a dictionary
        data = json.loads(json_string)
        # Return the value associated with the specified key
        return data.get(key, '')
    except json.JSONDecodeError:
        # Handle the case where the JSON decoding fails
        return ''

# Specify the key(s) you want to extract from the JSON strings
# Example: extracting 'displayName' and 'starRating'
keys_to_extract = ['reviewer.displayName', 'starRating']

# Create new columns for each key
for key in keys_to_extract:
    # Extract information using the defined function
    df[key.replace('.', '_')] = df['reviews'].apply(lambda x: extract_info(x, key))

# Display the resulting DataFrame
print(df)
```
